Import necesssary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Take initial angu;ar coordinates of the triangle as input from user

```python
#angular coordinates of the triangle
print("Given the following vector representation of the angular coordinates of a triangle, enter values for a, b and c \n X = [[a b+c]], \n Y = [[a b-c]] and \n Z = [[-a c]]\n")
a = float (input ("Enter value for a: "))
b = float (input ("Enter value for b: "))
c = float (input ("Enter value for c: "))

#take input from user
X = np.array([a, b+c])
Y = np.array([a, b-c])
Z8 = np.array([-a, c])
print("\nAccording to given input angular coordinates of the triangle are:\n X = [[", a," ", b+c, "]], \n Y = [[", a, " ", b-c, "]] and \n Z = [[",-a," ", c, "]]\n")
```

Plot the traingle and the vectors representation of the angular coordinates

```python
plt.rcParams["figure.figsize"] = [7, 6]
plt.rcParams["figure.autolayout"] = True
data = np.array([X, Y, Z])
origin = np.array([[0, 0, 0], [0, 0, 0]])
v1 = plt.quiver(*origin, data[0, 0], data[0, 1], color=['black'], scale=30)
v2 = plt.quiver(*origin, data[1, 0], data[1, 1], color=['red'], scale=30)
v3 = plt.quiver(*origin, data[2, 0], data[2, 1], color=['green'], scale=30)
plt.quiverkey(v1, 0.045, 0.049, 2, 'vector X', coordinates='data')
plt.quiverkey(v2, 0.045, 0.042, 2, 'vector Y', coordinates='data')
plt.quiverkey(v3, 0.045, 0.034, 2, 'vector Z', coordinates='data')
plt.title("Vectors X, Y and Z representing angular coordinates of the triangle")
plt.savefig('Vectors_Fig.png')
plt.grid()
plt.show()
```

```python
#Generate line points
def line_gen(m,n):
  len =10
  x_mn = np.zeros((2,len))
  lam_1 = np.linspace(0,1,len)
  for i in range(len):
    temp1 = m + lam_1[i]*(n-m)
    x_mn[:,i]= temp1.T
  return x_mn

#Generating all lines
x_XY = line_gen(X, Y)
x_YZ = line_gen(Y, Z)
x_XZ = line_gen(X, Z)

#Plotting all lines
plt.plot(x_XY[0,:],x_XY[1,:],label='$XY$')
plt.plot(x_YZ[0,:],x_YZ[1,:],label='$YZ$')
plt.plot(x_XZ[0,:],x_XZ[1,:],label='$XZ$')

plt.plot(X[0], X[1], 'o')
plt.text(X[0] * (1+0.025), X[1]*(1), 'X')
plt.plot(Y[0], Y[1], 'o')
plt.text(Y[0] * (1+0.025), Y[1]*(1) , 'Y')
plt.plot(Z[0], Z[1], 'o')
plt.text(Z[0] * (1+0.07), Z[1]*(1) , 'Z')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc='best')
plt.grid()
plt.title("Triangle formed by angular coordinates")
plt.savefig('Triangle_Fig.png')
plt.show()
```

Compute two sides of the triangle viz. X-Y and Y-Z

```python
#calculate sides of the triangle
s1 = X-Y
s2 = Y-Z
```

Compute area of triangle

```python
Area= (1/2)*(np.linalg.norm(np.cross(s1, s2)))
print('Area of the Triangle is {:.2f} square units'.format(Area))
```
